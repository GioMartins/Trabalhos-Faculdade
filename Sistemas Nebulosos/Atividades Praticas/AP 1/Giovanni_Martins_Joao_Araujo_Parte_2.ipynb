### 3. Segmentação de Imagens por Região:

Conforme explicado no arquivo principal, foi realizada uma outra técnica de segmentação de imagens, afim de acelerar este processo. Abaixo são apresentados os resultados obtidos.

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# Load the RGB image
img = cv2.imread('ImagensReduzidas/photo001.jpg')

# Reshape the image into a 2D array
X = img.reshape(-1, 3)

# Apply the FCM algorithm to the image
U, centers = fcm(X, n_clusters=3)

# Color each region (cluster) with the tonality of the pixel that corresponds to the center of the region
segmented_img = np.zeros_like(X)
for i in range(X.shape[0]):
    cluster = np.argmax(U[i])
    segmented_img[i] = centers[cluster]
segmented_img = segmented_img.reshape(img.shape)

# Color the pixels that have the highest degree of compatibility (pertinence) with a given region with the tonality of the central pixel of that region
pertinence = np.max(U, axis=1)
pertinence_img = pertinence.reshape(img.shape[:2])
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if pertinence_img[i, j] > 0.8:
            cluster = np.argmax(U[i*img.shape[1]+j])
            img[i, j] = centers[cluster]

# Display the original image, segmented image
print('Imagem Original')
cv2_imshow(img)
print('Imagem segmentada')
cv2_imshow(segmented_img)
```
